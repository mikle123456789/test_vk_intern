```python
import sqlite3

class Document():
    def __init__(self, url:str, PubDate:int, FetchTime:int, Text:str, FirstFetchTime = None):
        self.url = url
        self.PubDate = PubDate
        self.FetchTime = FetchTime
        self.Text = Text
        self.FirstFetchTime = FirstFetchTime if FirstFetchTime is not None else FetchTime


class Doc_Process():
    def __init__(self, dbfilename = 'test.db'):
        self.file = dbfilename
        self.doc_fields = ('url', 'PubDate', 'FetchTime', 'Text', 'FirstFetchTime')
        connection = sqlite3.connect(dbfilename)
        cursor = connection.cursor()
        cursor.execute('CREATE TABLE IF NOT EXISTS Documents (Url TEXT PRIMARY KEY, PubDate INTEGER NOT NULL, FetchTime Integer NOT NULL, Text TEXT NOT NULL, FirstFetchTime INTEGER)')
        connection.commit()
        connection.close()
    def process_document(self, doc:Document):
        connection = sqlite3.connect(self.file)
        cursor = connection.cursor()
        cursor.execute(f'''
                SELECT * FROM Documents
                       WHERE url = {doc.url}
            ''')
        previous = cursor.fetchall()
        #print(previous)
        if len(previous) == 0:
            #print(f'INSERT INTO Documents ({', '.join(self.doc_fields)}) VALUES ({doc.url}, {doc.PubDate}, {doc.FetchTime}, \"{doc.Text}\", {doc.FirstFetchTime})')
            cursor.execute(f'INSERT INTO Documents ({', '.join(self.doc_fields)}) VALUES ({doc.url}, {doc.PubDate}, {doc.FetchTime}, \"{doc.Text}\", {doc.FirstFetchTime})')
            connection.commit()
            connection.close()
            return doc
        else:
            #Создать результирующий документ и обновить запись в бд
            previous = previous[0]
            previous = Document(previous[0], previous[1], previous[2], previous[3], previous[4])
            previous:Document
            url_new = previous.url
            FetchTimeNew = max(previous.FetchTime, doc.FetchTime)
            PubDate_new = previous.PubDate if previous.FirstFetchTime < doc.FetchTime else doc.PubDate
            TextNew = previous.Text if previous.FetchTime > doc.FetchTime else doc.Text
            FirstFetchTimeNew = min(previous.FirstFetchTime, doc.FetchTime)
            new_doc = Document(url_new, PubDate_new, FetchTimeNew, TextNew, FirstFetchTimeNew)
            cursor.execute('''UPDATE Documents 
                            SET PubDate = ?, FetchTime = ?, Text = ?, FirstFetchTime = ?
                           WHERE Url = ?''', (PubDate_new, FetchTimeNew, TextNew, FirstFetchTimeNew, url_new))
            connection.commit()
            connection.close()
            return new_doc
```

```python
def print_result(res:Document):
    print(f'url: {res.url}; PubDate: {res.PubDate}; FetchTime: {res.FetchTime}; Text: {res.Text}; FirstFetchTime: {res.FirstFetchTime}')
```

```python
import random
Process = Doc_Process('test.db')
for i in range(10):
    url = random.randint(0, 9) + 10 * i 
    count = random.randint(1, 15)
    for j in range(count):
        PubDate = random.randint(0, 100)
        FetchTime = random.randint(0, 100)
        Text = f'Документ {url}, время {FetchTime}'
        doc = Document(url, PubDate, FetchTime, Text)
        print('Документ:')
        print_result(doc)
        res = Process.process_document(doc)
        print('Результат:')
        print_result(res)
```
